```python
import tensorflow as tf
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import load_img

#!pip install opencv-python
import cv2
```

```python
import numpy as np
from keras.models import load_model
from keras.preprocessing import image
```

```python
model = tf.keras.models.load_model('model (4).h5')
model1 = tf.keras.models.load_model('model1 (4).h5')
```

```python
final_results={0:'male',1:'female'}
GR_dict={0:(0,255,0),1:(0,0,255)}
# rect_size = 4
```

```python
from flask import Flask, render_template, redirect, url_for, request
app = Flask(__name__)

@app.route("/")
def home():
    return render_template("index.html")

@app.route('/uploader', methods = ['GET', 'POST'])
def upload_file():
     if request.method == 'POST':
        f = request.files['file']
        image_path="./images/"+f.filename
        f.save(image_path)
        img=load_img(image_path,(48,48))
        img=img_to_array(img)
        
        img1=np.array(tf.image.rgb_to_grayscale(
            img, name=None
            ))/255

        resized = cv2.resize(img1,(48,48))
    #         normalized = resized/255.0
        reshaped=resized.reshape(1, 48, 48, 1)
        result = model.predict(reshaped)
        result1 = model1.predict(reshaped)
        #a=result+" "+result1
        #esult=str(np.round(result[0][0]))
        if(np.round(result[0][0])==0):
            result="Male"
        else:
            result="Female"
        result1=str(np.round(result1[0][0]))
        a="Your gender is "+result+" and age is "+result1
        return render_template("index.html",name=result,a=result1) 
    
    
@app.route('/camera', methods = ['GET', 'POST'])
def camera():
    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    cap = cv2.VideoCapture(0)
    sum1=0
    count=0

    while True:
        ret, img = cap.read()
        #gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(img,1.05,5)

        for x,y,w,h in faces:

            face_img = img[y:y+h,x:x+w ]
    #         resized = np.array(face_img,target_size=(128,128))
            img1=np.array(tf.image.rgb_to_grayscale(
            face_img, name=None
            ))/255

            resized = cv2.resize(img1,(48,48))
    #         normalized = resized/255.0
            reshaped=resized.reshape(1, 48, 48, 1)
            result = model.predict(reshaped)
            result1 = model1.predict(reshaped)

    #         label = np.argmax(result,axis=1)[0]

            if result < 0.5:
                cv2.rectangle(img,(x,y),(x+w,y+h),GR_dict[1],2)
                cv2.rectangle(img,(x,y-40),(x+w,y),GR_dict[1],-1)
                cv2.putText(img, final_results[0], (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)
                print(final_results[0],np.round(result1[0][0]))
                sum1+=result1[0][0]
                count+=1
                break

            else :
                cv2.rectangle(img,(x,y),(x+w,y+h),GR_dict[0],2)
                cv2.rectangle(img,(x,y-40),(x+w,y),GR_dict[0],-1)
                cv2.putText(img, final_results[1], (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)
                print(final_results[1],result1[0][0])
                sum1+=result1[0][0]
                count+=1
                break

        cv2.imshow('LIVE', img)
        key = cv2.waitKey(1)
        if key == 27: 
            cap.release()
            cv2.destroyAllWindows()
            break

    cap.release()
    cv2.destroyAllWindows()
    return render_template("index.html",a=sum1//count)

if __name__ == "__main__":
    app.run()
```
